```python
import pandas as pd
```

```python
students = pd.read_excel('202 Student Demographics.xlsx', sheet_name="Students")
months = pd.read_excel('202 Student Demographics.xlsx', sheet_name="Months")
```

```python
students
```

```python
# get out of jail
stripped_months = []
for elt in students["Birth month"]:
    stripped_months.append(elt.strip())
students["Birth month"] = stripped_months
students
```

```python
# nicer Pandas version
students["Birth month"] = students["Birth month"].str.strip()
```

```python
print(len(students))
```

```python
months.head()
```

# Look up month names (`join`)

```python
students_months = pd.merge(
    students, months, left_on="Birth month", right_on="Month Name",
    how='inner'
)
print(len(students_months))
students_months
```

```python
renamed = students_months.rename(columns={
    "Number": "Month",
    "# semesters completed at Calvin": "n_semesters",
    "Height (inches)": "height_inches"
})
renamed
```

# Query by birth month (`query`)

```python
before_september = renamed[renamed["Month"] < 9]
before_september
```

```python
renamed.query("Month < 9")
```

# Summarize by group (`aggregate`)

```python
renamed["n_semesters"]
```

```python
students_per_semester = renamed.groupby("n_semesters").size()
students_per_semester
```

```python
avg_height_by_month = renamed.groupby("Month")["height_inches"].mean()
avg_height_by_month
```

```python
renamed["height_inches"].groupby(renamed["Month"]).mean().to_frame().reset_index()
```
