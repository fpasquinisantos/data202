```python
def transform(data):
    X = data[['total_bill']].assign(constant=1)
    y = data['%']
    return X, y
X_train, y_train = transform(data)
```

```python
def predict(X, theta):
    return np.dot(X, theta)

plt.scatter(data['total_bill'], y_train)
plt.plot(data['total_bill'], predict(X_train, [-10 / 40, 20.]))
```

```python
def rmse_given_params(X, y, theta):
    predictions = predict(X, theta)
    residuals = y - predictions
    return np.mean(residuals ** 2) ** .5
rmse_given_params(X_train, y_train, [-.25, 20])
```
