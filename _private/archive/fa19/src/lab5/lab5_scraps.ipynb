# Overfitting

```python
data = pd.DataFrame({"x": np.linspace(0.0, 1., 20)})
data['y'] = 10 * np.sin(10 * data.x) + np.random.standard_normal(len(data.x))
plt.scatter(data['x'], data['y'])
```

```python
X = data[['x']]
metrics.r2_score(data.y, LinearRegression().fit(X, data.y).predict(X))
```

```python
def transform_poly(data):
    X = data[['x']].copy()
    for i in range(2, 50):
        X[f'x^{i}'] = data.x ** i
    return X
X = transform_poly(data)
```

```python
model = LinearRegression().fit(X, data.y)
metrics.r2_score(data.y, model.predict(X))
```
